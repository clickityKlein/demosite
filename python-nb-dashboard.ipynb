```python
# data import
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

rand1 = np.random.rand(10)
rand2 = np.random.rand(10)
rand3 = np.random.rand(10)

rand_dict = {'random_set1': rand1, 'random_set2': rand2, 'random_set3': rand3}

df_rand = pd.DataFrame(rand_dict)
```

## Row

```python
#| title: Dataframe
df_rand
```

```python
#| title: Dataframe Plot
plt.plot(df_rand)
```

Can we do markdown?

```python
plt.hist(df_rand)
plt.legend(labels = df_rand.keys())
```

```python
plot_input = "random_set1"
plt.plot(df_rand[plot_input])
```

```python
import ipywidgets as widgets
```

```python
sidebar_input = widgets.RadioButtons(
    options = ["random_set1", "random_set2", "random_set3"],
    description = "Choose Random Set:",
    disabled = False
)
```

```python
def plot_input(sidebar_input):
    plt.plot(df_rand[sidebar_input])
    plt.show()
```

```python
widgets.interactive(plot_input, sidebar_input = sidebar_input)
```
